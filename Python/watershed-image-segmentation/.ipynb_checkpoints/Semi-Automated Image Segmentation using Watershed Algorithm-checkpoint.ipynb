# Semi-Automated Image Segmentation using Watershed Algorithm

___

Build a function enabling the semi-automated segmentation of 2D images using the functionaity of the Watershed Algorithm.

## Computer Vision
### Empty black frame setup
**Here, we will create a function that generate an single black image that has the exact same dimension as the imported image.**

```python
def empty_img_gen(shape):
    return np.zeros(shape, dtype=np.uint8)
```

**Apply function to image to generate a new array named _segments_ (as it will be used to display the resulting Watershed segments later on).**

```python
segments = empty_img_gen(img.shape)
print('Shape of original image:\t', img_copy.shape)
print('Shape of empty frame:\t\t', segments.shape)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(segments)
plt.show()
```

### Mouse callback for color marker selection
**Here, we will create a callback function that allows the user the click on the image and draw a marker using the left click, and switch color using the right click.**

To make sure that the function goes back to the first available color of the selected colormap when the user has passed all available colors, we will first create a dictionary that contains the names of all matplotlib qualitative colormaps as _keys_ and their respective number of colors as _values_.

_For more details on matplotlib colormaps, visit https://matplotlib.org/examples/color/colormaps_reference.html_

```python
cmap_length = {'Pastel1': 9, 'Pastel2': 8, 'Paired': 12, 'Accent': 8,
               'Dark2': 8, 'Set1': 9, 'Set2': 8, 'Set3': 12,
               'tab10': 10, 'tab20': 20, 'tab20b': 20, 'tab20c': 20}
```

**Select colormap.**

```python
cmap = 'Pastel1'
```

**Define total number of markers based on selected colormap length.**

```python
num_markers = cmap_length[cmap]
```

**Initialize clicked booleen as False (will turn to True when left click, so the algorithm knows when to look for segments).**

```python
left_click = False
right_click = False
```

**Define mouse callback function.**

```python
def mouse_callback(event, x, y, flags, param):
    
    global color_idx, left_click, right_click

    # Reset color index if all colors have been passed
    if right_click == True and event == cv2.EVENT_RBUTTONDOWN and color_idx == cmap_length[cmap]-1:
        right_click == False
        color_idx = 0
        # Generate RGB color code from current color index as a tuple
        exec('color_tup = tuple(np.array(cm.' + cmap + '(color_idx))[:3]*255)')
    
    # Switch color index based on right click
    elif right_click == True and event == cv2.EVENT_RBUTTONDOWN:
        right_click == False
        color_idx += 1  
        # Generate RGB color code from current color index as a tuple
        exec('color_tup = tuple(np.array(cm.' + cmap + '(color_idx))[:3]*255)')
        
    elif event == cv2.EVENT_LBUTTONDOWN:
        
        # Tracking for markers
        cv2.circle(marker_image, (x, y), 5, (color_idx), -1)
        
        # Display marker on image copy
        cv2.circle(img_resized, (x, y), 5, color_idx, -1)
        
        # Set clicked to True
        left_clicked = True
```

### Watershed algorithm implementation

```python
# Import random image and resize it
rand_idx = np.random.randint(0, len(data))
img = cv2.imread('data/' + data[rand_idx])

# Create a copy of the image
img_copy = np.copy(img)

# Resize image (so both the original and segmented images can be seen in full on a single laptop screen)
resize_ratio = 15 # percent of original size
width = int(img.shape[1] * resize_ratio / 100)
height = int(img.shape[0] * resize_ratio / 100)
dim = (width, height)
img_resized = cv2.resize(img_copy, dim, interpolation = cv2.INTER_AREA)

# Generate empty segments image
segments = empty_img_gen(img_resized.shape)

# Initialize marker tracking so the Watershed algorithm knows what markers belong to a segment
marker_image = np.zeros(img_resized.shape[:2], dtype=np.int32)

# Initialize color index
color_idx = 1

# Link mouse callback function to window
cv2.namedWindow('Image')
cv2.setMouseCallback('Image', mouse_callback)

while True:
    
    # Show the original image and the resulting Watershed segments in two separate windows
    cv2.imshow('Segmented Image', segments)
    cv2.imshow('Image', img_resized)        
        
    # Close everything if Esc is pressed
    k = cv2.waitKey(1)

    if k == 27:
        break
        
    # Reset everything if 'r' is pressed
    elif k == ord('r'):
        img_copy = img.copy()
        img_resized = cv2.resize(img_copy, dim, interpolation = cv2.INTER_AREA)
        marker_image = np.zeros(img_resized.shape[0:2], dtype=np.int32)
        segments = np.zeros(img_resized.shape, dtype=np.uint8)
        color_idx = 0
           
    # If we clicked somewhere, call the Watershed algorithm on our chosen markers
    if left_click:
        
        marker_image_copy = marker_image.copy()
        cv2.watershed(img_resized, marker_image_copy)
        
        segments = np.zeros(img_resized.shape, dtype=np.uint8)
        
        for color_idx in range(num_markers):
            # Generate RGB color code from current color index as a tuple
            exec('color_tup = tuple(np.array(cm.' + cmap + '(color_idx))[:3]*255)')
            segments[marker_image_copy == color_tup] = color_tup
        
        left_click = False
        
cv2.destroyAllWindows()
```
